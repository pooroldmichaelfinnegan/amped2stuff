```python
with open('/Users/g/_m/git/learnding/game_stuff/amped/BK1.rl', 'rb') as pgf:
    pgf = pgf.read()

from struct import unpack
```

```python
# bl = pgf[0x3C4F10:0x5CF120] # nz pgf 1C grouping
# bl = pgf[0x3D0888:0x5B5958] # bk2 pgf
# bl = pgf[0x1082C0:0x19a030] # bk1 trees pgf
bl = pgf[:0x342C]

l = []
group = 0x24
listt = [5,6]

for i, b1 in enumerate(bl[::group]):
    i *= group; ll = []
    for j in listt:
        j *= 4; b4 = bl[i+j:i+j+4]
        # print(b4)
        f = float(f"{unpack('f', b4)[0]:.6f}")
        ll += [f]
    l += [ll]


print(l)
```
